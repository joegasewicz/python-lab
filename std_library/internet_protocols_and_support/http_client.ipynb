```python
# http.client - HTTP protocol client


## parse_headers
from http.client import parse_headers
from io import BytesIO

# BytesIO lets you treat bytes in memory like a file
# which is required by http.client.parse_headers()

headers = BytesIO(b"Connection: keep-alive\r\nCache-Control: max-age=0\r\nsec-ch-ua: somthing\r\n\r\n")
parsed_headers = parse_headers(headers)
conn = parsed_headers.get("Connection")
conn

```
